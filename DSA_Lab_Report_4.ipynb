```python
import copy

# ===== Base Class =====
class GroceryItem:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        print(f"[Constructor] Created {self.__class__.__name__} - {self.name}")

    def __del__(self):
        print(f"[Destructor] Deleted {self.__class__.__name__} - {self.name}")

    # Mimic copy constructor
    def copy_constructor(self):
        print(f"[Copy Constructor] Cloning {self.name}")
        return copy.deepcopy(self)

    # Mimic copy assignment
    def assign(self, other):
        print(f"[Copy Assignment] Assigning {other.name} to {self.name}")
        self.name = other.name
        self.price = other.price
        self.quantity = other.quantity

    def calculate_sum(self):
        return self.price * self.quantity

    def display(self):
        print(f"{self.__class__.__name__}: {self.name} | Price: {self.price:.2f} | "
              f"Qty: {self.quantity} | Subtotal: {self.calculate_sum():.2f}")


# ===== Derived Classes =====
class Fruit(GroceryItem):
    pass

class Vegetable(GroceryItem):
    pass


# ===== Utility Functions =====
def TotalSum(grocery_list):
    return sum(item.calculate_sum() for item in grocery_list)

def PrintAll(grocery_list, title):
    print(f"\n=== {title} ===")
    for item in grocery_list:
        item.display()
    print(f"Total: {TotalSum(grocery_list):.2f}")


# ===== Driver Code =====
if __name__ == "__main__":
    # Problem 1: Create objects
    apples = Fruit("Apples", 20, 4)
    bananas = Fruit("Bananas", 5, 6)
    carrots = Vegetable("Carrots", 4, 10)
    lettuce = Vegetable("Lettuce", 10, 1)

    # Demonstrate copy constructor & assignment
    apple_copy = apples.copy_constructor()
    temp_fruit = Fruit("TempFruit", 0, 0)
    temp_fruit.assign(bananas)

    carrot_copy = carrots.copy_constructor()
    temp_veg = Vegetable("TempVeg", 0, 0)
    temp_veg.assign(lettuce)

    # Problem 2: GroceryList array
    GroceryList = [apples, bananas, carrots, lettuce]
    PrintAll(GroceryList, "Jenna's Grocery List (Initial)")

    # Problem 3: TotalSum is shown above
    print(f"Computed TotalSum: {TotalSum(GroceryList):.2f}")

    # Problem 4: Delete Lettuce
    for item in GroceryList:
        if item.name == "Lettuce":
            print(f"\nDeleting: {item.name}")
            GroceryList.remove(item)
            del item
            break

    PrintAll(GroceryList, "After Deleting Lettuce")
```
