```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def sort_array(arr):

    sorted_arr = sorted(arr)
    print("Sorted array:", sorted_arr)

def reverse(arr):
    reversed_arr = sorted(arr)[::-1]
    print("Reversed array:", reversed_arr)

def main():
    # Initialize an array
    arr = [30, 20, 50, 10, 40]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Sort array")
        print("3. Reverse Sorted array")
        print("4. Insert new element")
        print("5. Delete element")
        print("6. Modify element")
        print("7. Exit")
        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            sort_array(arr)
        elif choice == '3':
            reverse(arr)
        elif choice == '4':
            new_element = int(input("Enter the new element: "))
            arr.append(new_element)
            print("New element added:", new_element)
            print("Updated array:", arr)
        elif choice == '5':
            index_to_delete = int(input("Enter the index to delete: "))
            if 0 <= index_to_delete < len(arr):
                del arr[index_to_delete]
                print("Element at index", index_to_delete, "deleted.")
                print("Updated array:", arr)
            else:
                print("Invalid index. No element deleted.")
        elif choice == '6':
            index_to_modify = int(input("Enter the index to modify: "))
            if 0 <= index_to_modify < len(arr):
                new_value = int(input("Enter the new value: "))
                arr[index_to_modify] = new_value
                print("Element at index", index_to_modify, "modified to", new_value)
                print("Updated array:", arr)
            else:
                print("Invalid index. No element modified.")
        elif choice == '7':
            break
        else:
            print("Exiting the program.")
            print("Invalid choice. Please enter a number between 1 to 3.")

if __name__ == "__main__":
    main()
```
